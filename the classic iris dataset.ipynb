# What is our baseline ?

What is the simpliest model we can think of?
In our case, if we just randomly guess we would expect a accuracry of .33 since we have 3 different classes that are evenly balanced

Modeling- simple manual model
Lets manually look at our data and pick some classification

```python
def single_feature_prediction(petal_length):
    if petal_length<2.5:
        return 0
    elif petal_length < 4.8:
        return 1
    else:
        return 2
```

```python
df_train.columns
```

# where are we misclassifying points

```python
cvPredict= cross_val_predict(model,X_train,Y_train,cv=5)
```

```python
predicted_correctly_mask= cvPredict ==Y_train
```

```python
predicted_correctly_mask
```

```python
wrongpredictions=~predicted_correctly_mask
```

```python
X_train[wrongpredictions]
```

```python
df_predictions= df_train.copy()
```

```python
def plot_incorrect_predictions(df_predictions, x_axis_feature,y_axis_feature):
    fig,axs = plt.subplots(2,2,figsize=(10,10))
    axs=axs.flatten()
    sns.scatterplot(x=x_axis_feature,y=y_axis_feature,hue="prediction_label",data=df_predictions,ax=axs[0])
    sns.scatterplot(x=x_axis_feature,y=y_axis_feature,hue="prediction_label",data=df_predictions,ax=axs[1])
    sns.scatterplot(x=x_axis_feature,y=y_axis_feature,hue="correct_predicition",data=df_predictions,ax=axs[2])
    axs[3].set_visible(False)

```

```python
plot_incorrect_predictions(df_predictions,"petal length (cm)", "petal width (cm)")
```
